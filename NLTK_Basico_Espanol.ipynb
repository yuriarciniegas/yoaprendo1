**1. Text Data**

```
text_data = "Python es un lenguaje de programacin interpretado cuya filosofa hace hincapi en la legibilidad de su cdigo. " \
            "Se trata de un lenguaje de programacin multiparadigma, ya que soporta parcialmente la orientacin a objetos, " \
            "programacin imperativa y, en menor medida, programacin funcional. Es un lenguaje interpretado, dinmico y multiplataforma."
```

**2. Preprocess: Convert to LowerCase and Remove Special Characters**

```
from string import punctuation

preproc_data = text_data.lower()
#print(preproc_data)

#Remove numbers and punctuation symbols
preproc_data = ''.join(c for c in preproc_data if not c.isdigit())
preproc_data = ''.join(c for c in preproc_data if c not in punctuation)

print(preproc_data)
```

**3. Tokenization**

3A. Tokenization by sentences

```
import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt')

text_sent = sent_tokenize(text_data)

print(text_sent)
```

3B. Tokenization by words

```
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

tokens = tokenizer.tokenize(preproc_data)

print(tokens)
```

3C. Tokenize by bigrams

```
bigrm = nltk.bigrams(tokens)

for b in bigrm:
  print(b)
```

**4. Remove Stop Words**

4A. Download Spanish Lexicon

```
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('spanish'))
other_stop_words = set(['me', 'le', 'da', 'mi', 'su', 'ha', 'he', 'ya', 'un', 'una', 'es','del', 'las', 'los', 'en', 'que', 'y', 'la','de'])

stop_words = stop_words.union(other_stop_words)

print(stop_words)
```

4B. Remove Stop Words

```
filtered_words = [w for w in tokens if not w in stop_words]

print(filtered_words)
```

**5. Stemmize Text**

```
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer('spanish')

stemmed_text = [stemmer.stem(i) for i in filtered_words]

print(stemmed_text)
```

**6. Word Frequency**

```
fdist = nltk.FreqDist(stemmed_text)
for word, frequency in fdist.most_common(50):
    print(u'{}:{}'.format(word, frequency))

fdist.plot(30,cumulative=False)
```
