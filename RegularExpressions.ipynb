**Python Regular Expressions Package**

```
import re
```

**Basic functions**

```
text_base = "Mi mama me ama y mi papa me ama"
ex1 = re.search('ama', text_base)
print(ex1)
ex2 = re.findall('ama', text_base)
print(ex2)
ex3 = re.split('ama', text_base)
print(ex3)
ex4 = re.sub('ama', 'X', text_base)
print(ex4)
```

**Regular Expressions**

```
text_base = "Hola, me llamo Fran, mi correo es: fran.romero@gmail.com. " \
"Mi nmero de cdula es: 79965159. Tambien tengo otro correo y es fromeroa@yahoo.es. " \
"Leo las noticias en http://www.eltiempo.com y tambien me gusta visitar facebook en " \
"http://www.facebook.com. Mi celular es 3006789527. Algunos hashtags de twitter son: " \
"#CienciaDeDatos e #TextMining. Hoy es 09/08/2021 y el ao termina el da 31/12/2021. " \
"Hace poco anduve 45.67 km en bicicleta, logrando disminuir -0.1 kg. y peso .65 "

#regex = "\\d{2}/\\d{2}/\\d{4}"
#regex = "[-+]?(?:\d+(?:\.\d*)?|\.\d+)"
regex = "#[\\d|\\w]+"
match = re.findall(regex, text_base)
print(match)
```

**Simple Url Matcher**

```
regex = r'(http|ftp|https)(://)([\w\-_]+(?:(?:\.[\w\-_]+)+))([\w\-\.,@?^=%&amp;:/~\+#]*[\w\-\@?^=%&amp;/~\+#])?'
print(["".join(x) for x in re.findall(regex, text_base)])
```
